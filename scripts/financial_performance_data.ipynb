```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
aapl = yf.Ticker("AAPL")
income = aapl.financials
income
revenue = income.loc["Total Revenue"]
revenue
prices = aapl.history(period="5y")
prices.head()
prices["Close"].plot(figsize=(10,5))
plt.title("Apple Stock Price (Last 5 Years)")
plt.show()
```

```python
net_income = income.loc["Net Income"]
profit_margin = net_income / revenue
profit_margin
df = pd.DataFrame({
    "Revenue": revenue,
    "Net_Income": net_income,
    "Profit_Margin": profit_margin
})

df
```
